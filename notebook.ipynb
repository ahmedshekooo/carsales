### ---------------- MAKING INSIGHTS ----------------

##### --------- INSIGHT 1

```python
month_sales = {}
for i in range(1,13):
    month_sales[i] = df[df["MONTH_ID"] == i]["SALES"].sum()
print(month_sales)
```

```python
import calendar

# Get month names in correct order
month_names = [calendar.month_name[i] for i in range(1, 13)]
month_values = [month_sales[i] for i in range(1, 13)]

fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('#151515')
ax.set_facecolor('#151515')
ax.plot(month_names, month_values, marker='o', color='#ffda22', linewidth=3, alpha=0.8)
ax.set_xlabel("Month", fontsize=12, color='white')
ax.set_ylabel("Sales", fontsize=12, color='white')
ax.set_title("Sales by Month", fontsize=14, color='white')
ax.tick_params(colors='white', axis='x', rotation=45)
ax.tick_params(colors='white', axis='y')
ax.grid(axis='y', alpha=0.3, color='gray')
plt.tight_layout()
plt.show()
```

##### ---------------- INSIGHT 2

```python
productline_sales = {}
for i in df["PRODUCTLINE"].unique():
    productline_sales[i] = df[df["PRODUCTLINE"] == i]["SALES"].sum()
print(productline_sales)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
fig.patch.set_facecolor('#151515')
ax.set_facecolor('#151515')
wedges, texts, autotexts = ax.pie(
    list(productline_sales.values()),
    labels=list(productline_sales.keys()),
    autopct='%1.1f%%',
    colors=['#1f1f1f', '#ffda22', '#303030', '#4a4a4a', '#666666', '#878787', "#b3b3b3"],
    textprops={'color': 'white', 'fontsize': 14},
)
ax.set_title("Sales by Product Line", fontsize=16, color='white')
plt.tight_layout()
plt.show()
```

##### --------- INSIGHT 3

```python
country_sales = {}
for i in df["COUNTRY"].unique():
    country_sales[i] = df[df["COUNTRY"] == i]["SALES"].sum()

# Get top 10 countries by sales
top_10_countries = dict(sorted(country_sales.items(), key=lambda x: x[1], reverse=True)[:10])
print("Top 10 Countries by Sales:")
for country, sales in top_10_countries.items():
    print(f"{country}: ${sales:,.2f}")
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
fig.patch.set_facecolor('#151515')
ax.set_facecolor('#151515')
bars = ax.bar(top_10_countries.keys(), top_10_countries.values(), color='#ffda22', alpha=0.7, width=0.6)
ax.set_xlabel("Country", fontsize=12, color='white')
ax.set_ylabel("Sales", fontsize=12, color='white')
ax.set_title("Top 10 Countries by Sales", fontsize=14, color='white')
ax.tick_params(colors='white', axis='x', rotation=45)
ax.tick_params(colors='white', axis='y')
ax.grid(axis='y', alpha=0.3, color='gray')
plt.tight_layout()
plt.show()
```

##### --------- INSIGHT 4

```python
annual_sales = df.groupby('YEAR_ID')['SALES'].sum()

sales_2003 = annual_sales.get(2003, 0)
sales_2004 = annual_sales.get(2004, 0)
sales_2005 = annual_sales.get(2005, 0)

# Calculate Growth Rates
growth_03_04 = ((sales_2004 - sales_2003) / sales_2003) if sales_2003 > 0 else 0
growth_04_05 = ((sales_2005 - sales_2004) / sales_2004) if sales_2004 > 0 else 0

# Print results
print("--- Annual Sales Figures ---")
print(f"2003 Total Sales: ${sales_2003:,.2f}")
print(f"2004 Total Sales: ${sales_2004:,.2f}")
print(f"2005 Total Sales: ${sales_2005:,.2f}")

print("\n--- Sales Growth Rates ---")
print(f"2003 to 2004: {growth_03_04:.2%}")
print(f"2004 to 2005: {growth_04_05:.2%}")
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('#151515')
ax.set_facecolor('#151515')
ax.plot(annual_sales.index, annual_sales.values, marker='o', color='#ffda22', linewidth=3, alpha=0.8)
ax.set_xlabel("Year", fontsize=12, color='white')
ax.set_ylabel("Sales", fontsize=12, color='white')
ax.set_title("Annual Sales Trend", fontsize=14, color='white')
ax.tick_params(colors='white')
ax.grid(axis='y', alpha=0.3, color='gray')
plt.tight_layout()
plt.show()
```
